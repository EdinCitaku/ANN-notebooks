```python
# numpy
def gen_pattern(n_bits):
    array = np.random.randint(2, size=n_bits)
    array[array == 0] = -1
    return array

def gen_stored_patterns(n_patterns, n_bits):
    array = np.random.randint(2, size=n_bits*n_patterns)
    array[array == 0] = -1
    return array.reshape((n_patterns, n_bits))

def get_random_pattern(stored_patterns):
    idx = np.random.randint(stored_patterns.shape[0])
    return np.array(stored_patterns[idx])

def gen_weights(stored_patterns, n_bits = 0, diag_null = True):
    if n_bits == 0:
        n_bits = int(stored_patterns.size / stored_patterns.shape[0])
    weight_matrix = np.zeros(shape=(n_bits, n_bits), dtype=np.float64)
    for pattern in stored_patterns:
        weights = pattern.reshape((n_bits, 1)) @ pattern.reshape((1, n_bits))
        weight_matrix = weight_matrix + weights
    if diag_null:
            np.fill_diagonal(weight_matrix, 0)
    return np.true_divide(weight_matrix, n_bits)

def update_neuron(og_pattern, weight_matrix, index):
    weight_row = weight_matrix[index]
    new_neuron = np.sign(weight_row @ og_pattern)
    new_pattern = np.copy(og_pattern)
    new_pattern[new_pattern == 0] = 1
    new_pattern[index] = new_neuron
    return new_pattern
```

```python
def reco_digits(stored_patterns, og_pattern):
    weights_matrix = gen_weights(stored_patterns)
    c = 0
    og_pattern = og_pattern.reshape((160,1))
    new_pattern = np.copy(og_pattern)
    for i in range(new_pattern.size):
        new_pattern = update_neuron(new_pattern, weights_matrix, i)
    while not np.array_equal(new_pattern, og_pattern):
        clear_output(wait=True)
        display("{} iterations done.".format(c))
        c += 1
        og_pattern = np.copy(new_pattern)
        for i in range(new_pattern.size):
            new_pattern = update_neuron(new_pattern, weights_matrix, i)
    return new_pattern.reshape((16,10))
```
